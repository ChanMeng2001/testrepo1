```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
#read data
df = pd.read_csv("owid-covid-data.csv")
# To display the top 5 rows 
df.head(5)
```

```python
# To display the botton 5 rows
df.tail(5)
```

```python
#Checking the types of data
df.dtypes
```

```python
#Dropping irrelevant columns
df = df.drop(['excess_mortality_cumulative_per_million','excess_mortality','excess_mortality_cumulative','excess_mortality_cumulative_absolute','human_development_index','hospital_beds_per_thousand','handwashing_facilities','new_deaths_smoothed','new_cases_smoothed'], axis=1)
df.head(6)
```

```python
#Renaming the columns
df = df.rename(columns={'location':'country', 'total_cases':'TotalCases'})
```

```python
df.head(5)
```

```python
#Dropping the duplicate rows
df.shape #to check i have how many rows
```

```python
DuplicateRows = df[df.duplicated()]
print("Num of duplicated row = ", DuplicateRows.shape)
```

```python
df.count()      # Used to count the number of rows, only calculate row with full information, if a row has null then no count
```

```python
#Dropping the missing or null values.
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
#Detecting Outliers
df.head(5)
sns.boxplot(x=df('total_deaths'))
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Plot different features against one another (scatter), against frequency (histogram)
#histogram
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Death Rate")
plt.ylabel('total_deaths')
plt.xlabel('TotalCases');
#Heat Maps
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
#Scatterplot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['TotalCases'], df['total_deaths'])
ax.set_xlabel('TotalCases')
ax.set_ylabel('total_deaths')
plt.show()
```
